```python
import numpy as np
import pandas as pd
```

DataFrames is the combination of many columns.
It should have more than one column and more than one row so we can call it as a DataFrame.

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

Series - if we have only one row or only one column so it is a series

```python
df = pd.DataFrame(np.arange(0,20).reshape(5,4), index=["row1","row2","row3","row4","row5"],columns=['Col1','Col2','Col3','Col4']) #(Data, Index, Column)
```

```python
df.head()
```

```python
# Accessing the elements
# 1. loc
# 2. iloc
df.loc['row1']
```

```python
type(df.loc['row1'])
```

```python
df.iloc[2:4,1:3] # Here we can try many combinations as we want between rows and columns
```

```python
type(df.iloc[2:4,1:3]) # it will be dataframe since it contains more than one row an col
```

```python
print(df.iloc[0,:])
print(type(df.iloc[0:3,0])) # It is a series as it contains only one column
```

```python
## Specifying columns data types
data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')
```

```python
print(data)
```

```python
df = pd.read_csv(StringIO(data),dtype=object)
df
```

```python
df['a']
```

```python
df = pd.read_csv(StringIO(data),dtype=int) # Changing the datatype
df
```

```python
df['a'] # see the datatype
```

```python
df = pd.read_csv(StringIO(data),dtype=float) # Changing the datatype
df
```

```python
#Changing the datatype of particular columns
df = pd.read_csv(StringIO(data), dtype={'b':int, 'c':float, 'a':'int64','d':float})

df
```

```python
print(df['b'][1])
print(type(df['b'][1]))
```
